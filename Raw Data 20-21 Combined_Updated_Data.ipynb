```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import psycopg2 
import io
```

```python
VAERS_DATA_21_df = pd.read_csv('Resources/Updated_Data/2021VAERSDATA.csv') 
```

```python
VAERS_DATA_21_df.head()
```

```python
VAERS_DATA_20_df = pd.read_csv('Resources/Updated_Data/2020VAERSDATA.csv') 
VAERS_DATA_20_df.head()
```

```python
VAERS_DATA_ALL_df = VAERS_DATA_21_df.append(VAERS_DATA_20_df)

VAERS_DATA_ALL_df.head()
```

```python
len(VAERS_DATA_ALL_df.index)
```

```python
VAERS_DATA_ALL_df.drop_duplicates(subset ="VAERS_ID",
                     keep = False, inplace = True)
```

```python
len(VAERS_DATA_ALL_df.index)
```

```python
VAERS_DATA_ALL_df['DIED']=VAERS_DATA_ALL_df['DIED'].fillna(0)
```

```python
len(VAERS_DATA_ALL_df.loc[VAERS_DATA_ALL_df['DIED'] == 'Y'])
```

```python
 
engine = create_engine('postgresql://postgres:UTB00tcamp@team-11.ch3wmigorcf0.us-east-2.rds.amazonaws.com:5432/postgres', pool_recycle=3600)

VAERS_DATA_ALL_df.head(0).to_sql('vaersdata_raw', engine, if_exists='replace',index=False) #drops old table and creates new empty table

conn = engine.raw_connection()
cur = conn.cursor()
output = io.StringIO()
VAERS_DATA_ALL_df.to_csv(output, sep='\t', header=False, index=False)
output.seek(0)
contents = output.getvalue()
cur.copy_from(output, 'vaersdata_raw', null="") # null values become ''
conn.commit()
```
